```python
import numpy as np
from utils import Mean_Split
from ExKMCbundle.ExKMC.Tree import Tree
from ExKMCbundle.notebooks.utils import plot_kmeans, plot_tree_boundary
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
n = 100
d = 2
k = 3
```

```python
data, _ = make_blobs(n_samples=n, n_features=d, centers=k, cluster_std=1.0, random_state=0)
kmeans = KMeans(k)
kmeans.fit(data)
plot_kmeans(kmeans, data)
```

```python
ms = Mean_Split(data, dims=d, n_clusters=k)
ms.find_all_splits()
ms.plot_optimal_split()
```

```python
tree = Tree(k=k, max_leaves=2*k)
tree.fit_predict(data)
tree.plot('imgs/test')
```

```python
plot_tree_boundary(tree, k, data, kmeans)
```
