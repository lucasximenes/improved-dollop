```python
%matplotlib inline
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from ExKMC.Tree import Tree
from utils import calc_cost, plot_kmeans, plot_tree_boundary
```

```python
n = 1000
d = 2
k = 3
x_data, _ = make_blobs(n, d, k, 2.5, random_state=42)



kmeans = KMeans(k, random_state=42)
kmeans.fit(x_data)

plot_kmeans(kmeans, x_data)
```

```python
tree_1k = Tree(k)
tree_1k.fit(x_data, kmeans)

plot_tree_boundary(tree_1k, k, x_data, kmeans, plot_mistakes=True)
```

```python
tree_2k = Tree(k, max_leaves=2*k)
tree_2k.fit(x_data, kmeans)

plot_tree_boundary(tree_2k, k, x_data, kmeans, plot_mistakes=True)
```

```python
tree_n = Tree(k, max_leaves=n)
tree_n.fit(x_data, kmeans)

plot_tree_boundary(tree_n, k, x_data, kmeans, plot_mistakes=True)
```
